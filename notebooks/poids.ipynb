```python
# IMPORTS

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
from datetime import datetime
```

```python
# Define Constants
WEIGHT_DF_COLUMN_DATETIME='date'
WEIGHT_DF_COLUMN_WEIGHT='poids'
```

```python
# Loading weight data
df_weight=pd.read_csv('data/poids.csv', sep=';', parse_dates=[WEIGHT_DF_COLUMN_DATETIME])
```

```python
# GRAPH 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html / Format Strings
plt.plot(df_weight[WEIGHT_DF_COLUMN_DATETIME],df_weight[WEIGHT_DF_COLUMN_WEIGHT], '-o')
plt.title('Poid en kg', pad=10)
#hourFormatter=mdates.DateFormatter('%Hh')
#plt.gca().xaxis.set_major_formatter(hourFormatter) 
```
